```python
# Вариант 9 — Задача 1
# Найти количество элементов, кратных k, и максимальный из них
# Ввод матрицы из файла, вывод результата в файл

# Открываем файл с входными данными
with open("Хвостиков_Павел_Павлович_ЗИТ252_vvod1.txt", "r") as file:
    # Считываем порядок матрицы
    n = int(file.readline())
    # Считываем число k
    k = int(file.readline())

    # Считываем матрицу
    A = []
    for _ in range(n):
        line = file.readline().strip().split()
        row = list(map(int, line))
        A.append(row)

# Ищем элементы, кратные k
multiples = []
for row in A:
    for elem in row:
        if elem % k == 0:
            multiples.append(elem)

# Готовим вывод
if multiples:
    result_count = len(multiples)
    result_max = max(multiples)
else:
    result_count = 0
    result_max = "Нет элементов"

# Записываем результат в файл
with open("Хвостиков_Павел_Павлович_ЗИТ252_vivod1.txt", "w") as file:
    file.write(f"Количество элементов, кратных {k}: {result_count}\n")
    file.write(f"Максимальный из кратных: {result_max}\n")
```

```python
# Вариант 9 — Задача 2
# Найти максимальный по модулю элемент матрицы.
# Удалить строку и столбец, на пересечении которых он стоит.
# Ввод матрицы из файла, вывод результата в файл.

# Чтение входного файла
with open("Хвостиков_Павел_Павлович_ЗИТ252_vvod2.txt", "r") as file:
    n = int(file.readline())

    A = []
    for _ in range(n):
        line = file.readline().strip().split()
        row = list(map(float, line))
        A.append(row)

# Поиск максимального по модулю элемента
max_val = abs(A[0][0])
max_i = 0
max_j = 0

for i in range(n):
    for j in range(n):
        if abs(A[i][j]) > max_val:
            max_val = abs(A[i][j])
            max_i = i
            max_j = j

# Формируем новую матрицу без строки max_i и столбца max_j
new_matrix = []
for i in range(n):
    if i == max_i:
        continue
    row = []
    for j in range(n):
        if j == max_j:
            continue
        row.append(A[i][j])
    new_matrix.append(row)

# Запись результата в файл
with open("Хвостиков_Павел_Павлович_ЗИТ252_vivod.txt", "w") as file:
    file.write(f"Максимальный по модулю элемент: {A[max_i][max_j]}\n")
    file.write(f"Удалена строка {max_i} и столбец {max_j}\n")
    file.write("Полученная матрица:\n")
    for row in new_matrix:
        file.write(" ".join(map(str, row)) + "\n")
```
