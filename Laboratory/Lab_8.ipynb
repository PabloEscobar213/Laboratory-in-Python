```python
# Вариант 1 — Задача 1
# Сумма и количество положительных элементов над главной диагональю

# Ввод размера матрицы
n = int(input("Введите размер матрицы N: "))

# Ввод матрицы
A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

# Сумма и количество положительных элементов над главной диагональю
s = 0
count = 0

for i in range(n):
    for j in range(i + 1, n):   # j > i — элементы над диагональю
        if A[i][j] > 0:
            s += A[i][j]
            count += 1

print("Сумма положительных над диагональю:", s)
print("Количество положительных над диагональю:", count)
```

```python
# Вариант 2 — Задача 1
# Проверка, является ли матрица магическим квадратом

n = int(input("Введите размер матрицы: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

# Сумма первой строки — эталон
magic_sum = sum(A[0])
is_magic = True

# Проверяем строки
for row in A:
    if sum(row) != magic_sum:
        is_magic = False

# Проверяем столбцы
for j in range(n):
    col_sum = 0
    for i in range(n):
        col_sum += A[i][j]
    if col_sum != magic_sum:
        is_magic = False

print("Матрица магическая" if is_magic else "Матрица НЕ магическая")
```

```python
# Вариант 3 — Задача 1
# Проверка симметричности матрицы

n = int(input("Введите порядок матрицы: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

symmetric = True

for i in range(n):
    for j in range(n):
        if A[i][j] != A[j][i]:
            symmetric = False

print("Матрица симметрична" if symmetric else "Матрица НЕ симметрична")
```

```python
# Вариант 4 — Задача 1
# Найти строки с максимальной и минимальной суммой

m = int(input("Введите количество строк: "))
n = int(input("Введите количество столбцов: "))

A = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

# Считаем суммы строк
sums = [sum(row) for row in A]

max_row = sums.index(max(sums))
min_row = sums.index(min(sums))

print("Строка с максимальной суммой:", A[max_row], "Сумма =", sums[max_row])
print("Строка с минимальной суммой:", A[min_row], "Сумма =", sums[min_row])
```

```python
# Вариант 5 — Задача 1
# Упорядочить строки матрицы по возрастанию

m = int(input("Введите количество строк: "))
n = int(input("Введите количество столбцов: "))

A = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

# Сортировка каждой строки
for i in range(m):
    A[i].sort()

print("Матрица после сортировки строк:")
for row in A:
    print(*row)
```

```python
# Вариант 6 — Задача 1
# Максимумы строк и минимумы столбцов

n = int(input("Введите размер матрицы: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

# Максимумы строк
print("Максимумы строк:")
for row in A:
    print(max(row))

# Минимумы столбцов
print("Минимумы столбцов:")
for j in range(n):
    col = [A[i][j] for i in range(n)]
    print(min(col))
```

```python
# Вариант 7 — Задача 1
# Восстановить симметричную матрицу по верхнему треугольнику

n = int(input("Введите размер матрицы: "))

# Ввод верхнего треугольника
upper = []
for i in range(n):
    row = []
    for j in range(i, n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    upper.append(row)

# Восстановление полной матрицы
A = [[0]*n for _ in range(n)]

for i in range(n):
    for j in range(i, n):
        A[i][j] = upper[i][j - i]
        A[j][i] = upper[i][j - i]

print("Восстановленная матрица:")
for row in A:
    print(*row)
```

```python
# Вариант 8 — Задача 1
# Разделить элементы k-й строки на диагональный элемент

n = int(input("Введите порядок матрицы: "))
k = int(input("Введите номер строки k: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(float(input(f"A[{i}][{j}] = ")))
    A.append(row)

diag = A[k][k]  # диагональный элемент

for j in range(n):
    A[k][j] /= diag

print("Матрица после преобразования:")
for row in A:
    print(*row)
```

```python
# Вариант 9 — Задача 1
# Найти количество элементов, кратных k, и максимальный из них

n = int(input("Введите размер матрицы: "))
k = int(input("Введите число k: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

multiples = []

for row in A:
    for elem in row:
        if elem % k == 0:
            multiples.append(elem)

print("Количество кратных k:", len(multiples))
print("Максимальный из кратных:", max(multiples) if multiples else "Нет таких элементов")
```

```python
# Вариант 10 — Задача 1
# Максимальный элемент среди упорядоченных строк

m = int(input("Введите количество строк: "))
n = int(input("Введите количество столбцов: "))

A = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

candidates = []

for row in A:
    if row == sorted(row) or row == sorted(row, reverse=True):
        candidates.append(max(row))

if candidates:
    print("Максимальный среди упорядоченных строк:", max(candidates))
else:
    print("Нет упорядоченных строк")
```

```python
# Вариант 12 — Задача 1
# Найти такие k, что k-я строка совпадает с k-м столбцом

n = int(input("Введите размер матрицы: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

matches = []

for k in range(n):
    col = [A[i][k] for i in range(n)]
    if A[k] == col:
        matches.append(k)

print("Совпадающие строки и столбцы:", matches if matches else "Нет совпадений")
```

```python
# Вариант 14 — Задача 1
# Переставить строку с максимальным диагональным элементом со строкой m

n = int(input("Введите размер матрицы: "))
m = int(input("Введите номер строки m: "))

A = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

# Ищем максимальный элемент на диагонали
max_diag = A[0][0]
max_row = 0

for i in range(n):
    if A[i][i] > max_diag:
        max_diag = A[i][i]
        max_row = i

# Меняем строки
A[m], A[max_row] = A[max_row], A[m]

print("Матрица после перестановки:")
for row in A:
    print(*row)
```

```python
# Вариант 15 — Задача 1
# Найти строки, содержащие элемент c, и умножить их на d

m = int(input("Введите количество строк: "))
n = int(input("Введите количество столбцов: "))

c = int(input("Введите число c: "))
d = int(input("Введите число d: "))

A = []
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input(f"A[{i}][{j}] = ")))
    A.append(row)

for i in range(m):
    if c in A[i]:
        A[i] = [x * d for x in A[i]]

print("Матрица после преобразования:")
for row in A:
    print(*row)
```
